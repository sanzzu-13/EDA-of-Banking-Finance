```python


# Define custom colors
custom_palette = ["blue", "gray"]  # Example custom colors

# Create a boxplot for 'age' with respect to 'gender' using custom colors
sns.boxplot(y='age', data=df, x='gender', palette=custom_palette)

# Add title and labels to the plot
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')

# Customize ticks and labels
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

# Show the plot
plt.show()
```

* The dataset shows contains more female data than male data.
* Within the dataset, females aged between 24 and 31 are the most active age group.
* Due to the leftward shift of the median line, the box plots for females shows right skewness.
* Although tha data shows males aged between 26 and 30 are comparatively more numerous, their numbers are still lower than those of females.
* Similarly, the box plots for males also demonstrate right skewness due to the leftward shift of the median line.
* Overall, the dataset shows that females are more active in fixed deposit investments compared to males

```python
# #this is heat map pearson correlation matrix
# corrmat = df_numeric.corr(numeric_only=True)
# f, ax = plt.subplots(figsize=(16, 12))
# sns.heatmap(corrmat, vmax=0.8, square=True);
# plt.title('Pearson Correlation Matrix Heatmap',fontsize=16)
# plt.show()
# df_coy = df.copy()
# selecting cols where the datatype is int64
selected_columns = [col for col in df.columns if df[col].dtype == 'int64']

# Creating new df with selected columns
new_df = df[selected_columns].copy()


print(new_df.head())
```

```python
# Generate Pearson correlation matrix
corrmat = new_df.corr()

# Set up the plot figure
plt.figure(figsize=(12, 10))

# Create the heatmap using seaborn
sns.heatmap(corrmat, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)

# Add title
plt.title('Pearson Correlation Matrix Heatmap', fontsize=16)

# Rotate y-axis labels for better readability
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

* The heatmap illustrates the correlation between fixed deposit and age.
* There exists a weak negative correlation (-0.26) between age and fixed deposit.
* where 1 is a perfect positive correlation and -1 is perfect negative correlation.
* If the correlation value between age and fixed deposit is 0, it indicates no correlation between these two attribute.
* indicating that age is not strongly related to fixed deposit activity.
